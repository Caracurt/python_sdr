```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H = 2.0 * (np.random.normal(0, 1, (16, 16)) + 1j * np.random.normal(0, 1, (16, 16)))
```

```python
H_pl = H
for p_idx in range(4):
    plt.figure(p_idx)
    plt.imshow(np.abs(H), cmap='viridis')
    plt.title(f'p_idx={p_idx}')
    H_pl = H * H_pl
```
